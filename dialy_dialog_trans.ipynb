```python
import time
import pandas as pd
import asyncio
import nest_asyncio
from googletrans import Translator

nest_asyncio.apply()

input_csv = "/home/user/ml_lab/vaders_sentiment.csv"
output_csv = "/home/user/ml_lab/translated_output.csv"
text_column = "statement"
new_column = "telugu_translation"

translator = Translator()

async def safe_translate_async(text, src="en", dest="te", retries=5, delay=2):
    if pd.isna(text) or not isinstance(text, str) or text.strip() == "":
        return ""
    for attempt in range(retries):
        try:
            t = await translator.translate(text, src=src, dest=dest)
            return t.text
        except Exception as e:
            print(f"âš ï¸ Error: {type(e).__name__} | Retry {attempt+1}/{retries} in {delay}s")
            await asyncio.sleep(delay)
    return text

def safe_translate_sync(text, src="en", dest="te", retries=5, delay=2):
    if pd.isna(text) or not isinstance(text, str) or text.strip() == "":
        return ""
    for attempt in range(retries):
        try:
            t = translator.translate(text, src=src, dest=dest)
            return t.text
        except Exception as e:
            print(f"âš ï¸ Error: {type(e).__name__} | Retry {attempt+1}/{retries} in {delay}s")
            time.sleep(delay)
    return text

async def main():
    df = pd.read_csv(
        input_csv,
        encoding="ISO-8859-1",
        on_bad_lines="skip",
        engine="python"
    )

    total = len(df)
    translated = []
    print(f"ğŸš€ Starting translation of {total} rows...")

    is_async = asyncio.iscoroutinefunction(translator.translate)

    for i, text in enumerate(df[text_column], 1):
        if is_async:
            trans = await safe_translate_async(text)
        else:
            trans = safe_translate_sync(text)

        translated.append(trans)
        print(f"âœ… Row {i}/{total} translated")
        await asyncio.sleep(0.5) if is_async else time.sleep(0.5)

    df[new_column] = translated
    df.to_csv(output_csv, index=False, encoding="utf-8-sig")
    print(f"\nâœ… Translation completed successfully and saved to '{output_csv}'")

await main()
```
